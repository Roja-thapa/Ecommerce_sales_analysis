```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default= "plotly_white"
```

```python
df= pd.read_csv("Sample - Superstore.csv",encoding='latin-1')
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().any()
```

```python
df.describe()
```

```python
df.info()
```

**Order date and ship date are in object datatype rather than date format.Therefore we need to change it's data type **

**CHANGE DATA TYPE FROM OBJECT TO DATETIME FORMAT**

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])
df.info()
```

**ADDING NEW COLUMNS**

```python
df['order month']=df['Order Date'].dt.month
df['order month_name']=df['Order Date'].dt.strftime('%B')  # monthname
df['order year']=df['Order Date'].dt.year
df['order Day of week']=df['Order Date'].dt.dayofweek
df.head()
```

```python
df.columns
```

# 1.MONTHLY SALES ANALYSIS

```python
monthly_sales=df.groupby('order month')['Sales'].sum().reset_index()
monthly_sales
```

```python
fig= px.line(monthly_sales,x='order month',y='Sales', title='Monthly sales analysis')
fig.show()
```

Through this graph, we can visualize that Febraury(2)
had the lowest sales at approximately $59.7k, while November(11)had the highest sales at around 352.4k.

 # 2.SALES BY CATEGORY AND DETERMINE WHICH CATEGORY HAD THE HIGHEST AND LOWEST SALES


```python
category_sales = df.groupby('Category')['Sales'].sum().reset_index()
category_sales
```

```python
fig=px.pie(category_sales,names='Category',values='Sales',title='Sales by Category',hole=0.2)
fig.update_traces(textposition='inside',textinfo='percent+label') #update_traces is used to modify the figure
fig.show()
```

This pie chart shows that out of the three product categories â€” Furniture, Technology, and Office Supplies â€” Technology made the most money, making up about 36.4% of total sales (around $836,154). On the other hand, Office Supplies made the least amount in sales.



# 3.SALES ANALYSIS NEEDS TO BE DONE BASED ON  SUB CATEGORIES

```python
subcategory_sales= df.groupby(['Category','Sub-Category'])['Sales'].sum().reset_index()
subcategory_sales
```

```python
fig=px.bar(subcategory_sales,x='Sub-Category',y='Sales',title= 'Sales by Sub-Category',color='Category')
fig.show()
```

# 4.ANALYZE THE MONTHLY PROFIT FROM SALES AND DETERMINE WHICH MONTH HAD THE HIGHEST PROFIT

```python
df.head()
```

```python
monthly_profit= df.groupby('order month')['Profit'].sum().reset_index()
monthly_profit
```

```python
fig= px.line(monthly_profit,x='order month',y='Profit',title='Monthly Profit Analysis')
fig.show()
```

# 5.ANALYZE THE PROFIT BY CATEGORY AND SUB CATEGORY

```python
profit_by_category= df.groupby('Category')['Profit'].sum().reset_index()
profit_by_category



```

```python
custom_colors = {
    'Furniture': 'green',
    'Technology': 'yellow',
    'Office Supplies': 'blue'}

fig= px.bar(profit_by_category,x='Category',y='Profit',title='Profit by Category',color='Category',
            color_discrete_map=custom_colors)  # Apply custom colors

fig.show()
```

```python
profit_by_subcategory= df.groupby('Sub-Category')['Profit'].sum().reset_index()
profit_by_subcategory
```

```python
fig= px.pie(profit_by_subcategory,names='Sub-Category',values='Profit',title='PROFIT BY SUB-CATEGORY')
fig.update_traces(textinfo='percent+label')
fig.show()
```

```python
# To find Copiers is located in which category
df[df['Sub-Category']=='Copiers']['Category'].value_counts()
```

# 6.SALES AND PROFIT - CUSTOMER SEGMENT

```python
#We need to consider sales, profit and segment columns
df.head()
```

```python
df['Segment'].value_counts()
```

```python
sp_segment= df.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
sp_segment
```

```python
fig= px.bar(sp_segment,x='Segment',y=['Sales','Profit'],barmode='group',title='Sales and Profit by Segment')
fig.show()
```

# 7.ANALYZE THE SALES TO PROFIT RATIO

```python
sp_ratio= df.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
sp_ratio['sales_profit_ratio']=sp_ratio['Sales']/sp_ratio['Profit']
sp_ratio
```
